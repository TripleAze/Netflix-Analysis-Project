```python

columns_to_drop = ['Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22','Unnamed: 22','Unnamed: 23','Unnamed: 24','Unnamed: 25']


df = df.drop(columns=columns_to_drop)

df.head(2)
```

```python
columns_to_drop2 = ['Unnamed: 12','Unnamed: 13','Unnamed: 14','Unnamed: 15']

df = df.drop(columns=columns_to_drop2)

df.head(2)
```

```python
df.isnull().value_counts()
```

```python
# Check for columns with missing values
missing_values = df.isnull().sum()

# Filter to show only columns with missing values
missing_columns = missing_values[missing_values > 0]
print(missing_columns)

#  Get only the column names with missing values
missing_columns_names = missing_columns.index.tolist()
print(missing_columns_names)
```

```python
df.head(2)
```

```python
df['type'] = df['type'].str.strip().str.lower().str.title()


df['listed_in'] = df['listed_in'].str.strip().str.title()

df
```

```python
release_year_data = df.groupby('release_year').size()

plt.figure(figsize=(16, 8))
plt.plot(release_year_data.index, release_year_data.values, marker='o')
plt.title("Number of Releases by Year")
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.grid(True)
plt.show()
```

```python
# Convert date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract month from date_added
df['month_added'] = df['date_added'].dt.month

# Group by month
monthly_data = df.groupby('month_added').size()

plt.figure(figsize=(16, 8))
plt.bar(monthly_data.index, monthly_data.values, color="skyblue")
plt.title("Number of Releases by Month")
plt.xlabel("Month")
plt.ylabel("Number of Additions")
plt.xticks(range(1, 13), 
           ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python


# Split directors by comma, then explode to get each director in a separate row
directors_series = df['director'].str.split(',').explode().str.strip()

# Get the top 10 most frequent directors
top_directors = directors_series.value_counts().head(10)

# Plot
plt.figure(figsize=(16, 8))
top_directors.plot(kind='bar', color='salmon')
plt.title("Top 10 Directors by Number of Titles")
plt.xlabel("Director")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

```python
# Fill missing cast values 
df['cast'] = df['cast'].fillna("Unknown")

# Split cast members by comma, then explode to get each actor in a separate row
cast_series = df['cast'].str.split(',').explode().str.strip()

# Get the top 10 most frequent cast members
top_cast = cast_series.value_counts().head(10)

# Plot
plt.figure(figsize=(16, 8))
top_cast.plot(kind='bar', color='blue')
plt.title("Top 10 Cast Members by Number of Appearances")
plt.xlabel("Cast Member")
plt.ylabel("Number of Appearances")
plt.xticks(rotation=40)
plt.show()

```

```python
# Filter for a specific genre, e.g., Drama 
genre_filtered = df[df['listed_in'].str.contains('Drama', na=False)]

# Repeat director or cast analysis on `genre_filtered`
directors_genre_series = genre_filtered['director'].str.split(',').explode().str.strip()
top_directors_genre = directors_genre_series.value_counts().head(10)

# Plot
plt.figure(figsize=(16, 8))
top_directors_genre.plot(kind='bar', color='lightgreen')
plt.title("Top Directors in Drama Genre by Number of Titles")
plt.xlabel("Director")
plt.ylabel("Number of Titles")
plt.xticks(rotation=40)
plt.show()
```

```python
# Drop rows where any cell has the value "Unknown"
df = df.replace("Unknown", pd.NA)  # Replace "Unknown" with actual NaN values
df = df.dropna()  # Drop rows with any NaN values

# Alternatively, if you want to specify specific columns, you could do:
# df = df[~df[['director', 'cast']].isin(['Unknown']).any(axis=1)]


```

```python
df.head()
```

```python


# Create a column for each genre
genre_columns = df['listed_in'].str.get_dummies(sep=', ')

# Combine with country
country_genre_data = pd.concat([df['country'], genre_columns], axis=1)

# Count the number of titles for each genre per country
country_genre_counts = country_genre_data.groupby('country').sum()

# Calculate the correlation
correlation_matrix = country_genre_counts.corr()

# Plotting the heatmap
plt.figure(figsize=(20, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title("Correlation between Genres and Countries")
plt.show()
```
